```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

train_path = "../data/train.csv"
test_path = "../data/test.csv"

df = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

df.head()
```

```python
df.shape, df_test.shape

df.info()

df.describe().T
```

```python
df["SalePrice"].hist(bins=50, figsize=(8,5))
plt.title("Distribution of SalePrice")
plt.savefig("plots/saleprice_distribution.png", dpi=120)
plt.show()
```

```python
df["LogSalePrice"] = np.log1p(df["SalePrice"])

df["LogSalePrice"].hist(bins=50, figsize=(8,5))
plt.title("Distribution of Log SalePrice")
plt.savefig("plots/saleprice_distribution.png", dpi=120)
plt.show()
```

```python
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False).head(20)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
no_object_columns = [
    "PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu",
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
    "MasVnrType"
]

for col in no_object_columns:
    df[col] = df[col].fillna("None")
```

```python
df["GarageYrBlt"] = df["GarageYrBlt"].fillna(0)
df["MasVnrArea"] = df["MasVnrArea"].fillna(0)
```

```python
df["LotFrontage"] = df.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode()[0])
```

```python
df.isnull().sum().sum()
```
